```python
import csv
import os
class employeemanager:
    def __init__(self,filename="employee.csv"):
        self.filename=filename
        self.employees={}
        self.loaddata=()
    def loaddata(self):
        file = open(self.filename, 'r', encoding='utf-8')
        reading=csv.DictReader(file)
        for r in reading:
            self.employees [r["ID"]]={
                "NAME":r["NAME"],
                "POSITION":r["POSITION"],
                "SALARY":r["SALARY"],
                "EMAIL":r["EMAIL"]}
        file.close()
    def savedata(self):
        file= open(self.filename,'w',encoding='utf-8')
        fields=["ID","NAME","POSITION","SALARY","EMAIL"]
        writing=csv.DictWriter(file,fieldnames=fields)
        writing.writeheader()
        for eID,data in self.employees.items():
            writing.writerow({"ID":eID,**data})
        file.close()
    def addemployee(self):
        eID=input("enter id ")
        if eID in self.employees:
            print("this id already exists")
        NAME=input("enter name")
        POSITION=input("enter position")
        while True:
            SALARY=input("enter salary")
            if SALARY.isdigit():
                break
                
            else:
                print("salary has to be numeric,try again")
        while True:
            EMAIL=input("enter email")
            if "@" in EMAIL:
                break
                
            else:
                print("wrong format, try again")
        self.employees[eID]={
            "NAME":NAME,
            "POSITION":POSITION,
            "SALARY":SALARY,
            "EMAIL":EMAIL}
        self.savedata()
    
    def viewemployee(self):
        for eID,data in self.employees.items():
            print(f"ID:{eID},NAME:{data['NAME']},POSITION:{data['POSITION']},SALARY:{data['SALARY']},EMAIL:{data['EMAIL']}")
    
    def updateemployee(self,**updates):
        eID=input("enter the id of the employee to update")
        if eID in self.employees:
            update=input("what do you want to update")
            if update.lower()=="name":
                name=input("enter the new name")
                self.employees[eID]['NAME']=name
            if update.lower()=="position":
                position=input("enter the new position")
                self.employees[eID]['POSITION']=position
            if update.lower()=="salary":
                salary=input("enter the new salary")
                self.employees[eID]['SALARY']=salary
            if update.lower()=="email":
                email=input("enter the new email")
                self.employees[eID]['EMAIL']=email
        
        else:
            print("not an existing id")   
        self.savedata()
        
    def deleteemployee(self):
        eID=input("enter the id of the employee to delete")
        if eID in self.employees:
            del self.employees[eID]
        else:
            print("not an existing id")
        self.savedata()
        
    
    def searchemployee(self):
        eID=input("enter the id of the employee to search")
        if eID in self.employees:
            data=self.employees[eID]
            print(f"ID:{eID},NAME:{data['NAME']},POSITION:{data['POSITION']},SALARY:{data['SALARY']},EMAIL:{data['EMAIL']}")
        else:
            print("not an existing id")
    
    def menu(self):
        while True:
            print("employee managment system")
            print("1.add")
            print("2.view")
            print("3.update")
            print("4.delete")
            print("5.search")
            print("6.exit")
            x=int(input("choose the number of action you need"))
            if x==1:
                self.addemployee()
            elif x==2:
                self.viewemployee()
            elif x==3:
                self.updateemployee()
            elif x==4:
                self.deleteemployee()
            elif x==5:
                self.searchemployee()
            elif x==6:
                self.savedata()
                break
            else:
                print("wrong number")
manager=employeemanager()
manager.menu()
        
```
